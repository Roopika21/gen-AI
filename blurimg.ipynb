```python
image = cv2.imread('rose.jpeg',cv2.IMREAD_GRAYSCALE)
_, binary_segmentation=cv2.threshold(image,127,255,cv2.THRESH_BINARY )
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title("original image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(binary_segmentation,cmap='gray')
plt.title("binary_segmentation")
plt.axis("off")
plt.show()
```

```python
image = cv2.imread('rose.jpeg',cv2.IMREAD_GRAYSCALE)
_, otsu=cv2.threshold(image,0,255,cv2.THRESH_BINARY +cv2.THRESH_OTSU)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(image,cmap='gray')
plt.title("original image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(otsu,cmap='gray')
plt.title("image segmentation with otsu threshold")
plt.axis("off")
plt.show()
```

```python
import cv2

# Load the images
image1 = cv2.imread('1.jpeg')
image2 = cv2.imread('2.jpeg')

# Check if images are loaded properly
if image1 is None or image2 is None:
    print("Error: Could not load one or both images.")
    exit()

# Ensure both images are the same size; resize if necessary
if image1.shape != image2.shape:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Combine the images with weighted sum
alpha = 0.6  # Weight for the first image
beta = 0.4   # Weight for the second image
weighted_sum = cv2.addWeighted(image1, alpha, image2, beta, 0)

# Display the resulting image
cv2.imshow('Weighted Image', weighted_sum)

# Wait for a key press and close the window
if cv2.waitKey(0) & 0xFF == 27:
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load images
image1 = cv2.imread('1.jpeg')
image2 = cv2.imread('2.jpeg')

# Check if images are loaded successfully
if image1 is None or image2 is None:
    print("Error: One of the images did not load.")
    exit()

# Resize image2 to match image1's dimensions
if image1.shape[:2] != image2.shape[:2]:
    image2 = cv2.resize(image2, (image1.shape[1], image1.shape[0]))

# Normalize weights
weight1 = 0.5
weight2 = 0.5

# Blend images
weightedSum = cv2.addWeighted(image1, weight1, image2, weight2, 0)

# Display the resulting image
cv2.imshow('Weighted Image', weightedSum)

# Wait for user to press a key, then close the window
cv2.waitKey(0)
cv2.destroyAllWindows()
```
